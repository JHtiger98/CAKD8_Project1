```python
import pandas as pd
import numpy as np
import json
import folium
import platform
import matplotlib.pyplot as plt
```

# folium 기반

```python
# 서울특별시 구 경계 데이터
geo_seoul= json.load(open('./data/seoul_geo_simple.json',encoding = "UTF-8"))
```

```python
children_pop = pd.read_csv("./data/어린이인구.csv", 
                           usecols = ['행정구역','2022년10월_계_연령구간인구수'])
children_pop.rename(columns = {'2022년10월_계_연령구간인구수':'pop'}, inplace=True)

# 행정구역 '구'이름만 저장
children_pop['행정구역'] = children_pop['행정구역'].str.split(' ').str[1]
children_pop['행정구역'][0] = '서울특별시'

# 옛날 행정구역 코드를 위해 인구 데이터 불러오기
seoul_pop = pd.read_csv("./data/Population_SIG.csv", nrows=26)

# code 합치고 서울특별시 제거
children_pop.insert(0,'code',seoul_pop['code'])
children_pop.drop([0],inplace=True)

# 코드 문자열로 전환/ 인구수 숫자로 변환
children_pop['code'] = children_pop['code'].astype('str')
children_pop['pop'] = children_pop['pop'].str.replace(',','')
children_pop['pop'] = children_pop['pop'].astype('int64')
```

```python
m = folium.Map(location=[37.55, 127], tiles='cartodbpositron', zoom_start=10)
m.choropleth(
    geo_data=geo_seoul,
    name='인구수',
    data=children_pop,
    columns=['행정구역', 'pop'],
    key_on='feature.properties.name',
    fill_color='Reds',   
    line_opacity=0.3,
    legend_name = 'clidren_Population')

folium.LayerControl().add_to(m)
m
```

# matplot 기반

```python
korea = pd.read_csv('./data/data_draw_korea.csv', index_col=0, encoding='UTF-8')
seoul = korea.query('광역시도 == "서울특별시"')
seoul = seoul.merge(children_pop, on='행정구역')
```

```python
seoul
```

```python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

whitelabelmin = (max(seoul['pop']) - min(seoul['pop'])) * 0.25 + min(seoul['pop'])


vmin = min(seoul['pop'])
vmax = max(seoul['pop'])

mapdata = seoul.pivot(index='y', columns='x', values='pop')
masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)

plt.figure(figsize=(8, 13))
plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap='Reds', edgecolor='#aaaaaa', linewidth=0.5)

# 지역 이름 표시
for idx, row in seoul.iterrows():
    annocolor = 'white' if row['pop'] > whitelabelmin else 'red'


    plt.annotate(row['행정구역'], (row['x']-3.5, row['y']-0.5), weight='bold',
                 ha='center', va='center', color=annocolor)
                 
    
# 시도 경계 그린다.

plt.gca().invert_yaxis()
plt.gca().set_aspect(1)

plt.axis('off')
    
cb = plt.colorbar(shrink=.1, aspect=10)
cb.set_label('pop')

plt.tight_layout()
plt.show()
```
